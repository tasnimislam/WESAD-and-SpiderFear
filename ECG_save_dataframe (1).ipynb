```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from glob import glob
import pickle
import re
import json
```

```python
DIRECTORY_JSON = '/content/drive/MyDrive/PhD research/Stress Detection/json save all data instance'
```

```python
def resumeAfter(fileName, df):
  files = glob(DIRECTORY_JSON + f'/{fileName}_*.json') # 4_class_1.json

  epochNumbers = []
  for fileNameB in files:
    numbers = re.findall(r'\d+', fileNameB)
    epochNumbers.append(int(numbers[-1]))

  lastUpdated = max(epochNumbers)

  index = lastUpdated*int(0.25*700)

  if index+int(60*700)<len(df): # check if it has to be updated more
    print(f'Done for this instance for filename: {fileName}')
    return

  else:
    return index
```

```python
def getLastFileName(fileName):
  fileName = fileName.split('/')[-1]
  fileName = fileName.split('.csv')[0]
  return fileName
```

```python
def dfWindowECG(df, fileName, resume):
  '''
  if resume = True: strat from where left
  '''
  index = 0
  counter = 0

  fileName = getLastFileName(fileName)

  if resume==True:
    index = resumeAfter(fileName, df)
    if index==None:
      return

  while(index+int(60*700)<len(df)):
    dict_json = {}
    dict_json['data'] = list(df['c_ecg'][index: index+index+int(60*700)])
    if int(df['w_label'][index])==3:
      dict_json['label'] = 2
    else:
      dict_json['label'] = int(df['w_label'][index])

    with open(f'/content/drive/MyDrive/PhD research/Stress Detection/json save all data instance/{fileName}_{counter}.json', 'w') as fp:
      json.dump(dict_json, fp)

    counter +=1
    index = index + int(0.25*700)
```
